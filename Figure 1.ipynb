```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("Data/SuicideDeathRates.csv")
```

```python
df2 = df.loc[:, ["UNIT_NUM", "STUB_LABEL", "YEAR", "ESTIMATE"]]

df2 = df2[(df2["UNIT_NUM"] == 1)&((df2["STUB_LABEL"] == "All persons")|
          (df2["STUB_LABEL"] == "Male")|(df2["STUB_LABEL"] == "Female"))]
```

```python
columns = df2["STUB_LABEL"].unique().tolist()

year_index = df2["YEAR"].unique().tolist()

all_persons = df2["ESTIMATE"][df2["STUB_LABEL"] == "All persons"].tolist()

male = df2["ESTIMATE"][df2["STUB_LABEL"] == "Male"].tolist()

female = df2["ESTIMATE"][df2["STUB_LABEL"] == "Female"].tolist()
```

```python
data = {"All persons": all_persons, "Male": male, "Female": female}

df3 = pd.DataFrame(data, index = year_index)
```

```python
plt.figure(figsize = (10,5))
plt.plot(df3)
plt.title("Deaths/100,000 residents (age-adjusted)", fontsize = 14)
plt.legend(columns, loc = "best")
plt.xlabel("Year", fontsize = 14)
plt.ylabel("Estimate", fontsize = 14)
plt.xticks(range(1950, 2019, 4))
plt.show()
```
